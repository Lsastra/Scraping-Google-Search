```python
from selenium import webdriver 
import pandas as pd 
from selenium.webdriver.support.ui import WebDriverWait 
```

```python
url = "https://www.google.com/search?q=asuransi+jiwa&safe=strict&source=lnms&tbm=nws&sa=X&ved=2ahUKEwiDi6WL1MjnAhVLzjgGHRipD8sQ_AUoAXoECA4QAw&biw=919&bih=750&dpr=1.25"
driver = webdriver.Chrome() 
driver.get(url)
```

```python
user_data = driver.find_elements_by_xpath('//*[@class="fl"]')
links = []
for i in user_data:
            links.append(i.get_attribute('href'))
```

```python
links = links[2:]
```

```python
wait = WebDriverWait(driver, 10)
judul = []
sumber = []
waktu = []
highlight = []
link = []
for x in links:
    driver.get(x)
    v_judul = driver.find_elements_by_xpath('//*[@class="l lLrAF"]')
    for berita in v_judul:
        link.append(berita.get_attribute('href'))
    v_sumber = driver.find_elements_by_xpath('//*[@class="xQ82C e8fRJf"]')
    v_waktu = driver.find_elements_by_xpath('//*[@class="f nsa fwzPFf"]')
    v_highlight = driver.find_elements_by_xpath('//*[@class="st"]')
    total = len(v_judul)
    for y in range(total):
        judul.append(str(v_judul[y].text))
        sumber.append(str(v_sumber[y].text))
        waktu.append(str(v_waktu[y].text))
        highlight.append(str(v_highlight[y].text))
```

```python
df = pd.DataFrame(set(zip(judul, sumber, waktu, highlight, link)))
df.columns = ["Judul", "Sumber", "Waktu", "Highlight", "link"]
```
